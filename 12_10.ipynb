# Damped Driven Coupled Oscillator

### Import packages

```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
init_printing()
dpi = 180
mpl.rcParams['figure.dpi'] = dpi
```

### Set up differential equation

```python
eta1, eta2 = symbols('eta_1 eta_2', cls=Function)
t, omega0, beta, omega, A, epsilon = symbols('t omega_0 beta omega A epsilon', real=True)

eta1_d = diff(eta1(t), t)
eta1_dd = diff(eta1_d, t)

eta2_d = diff(eta2(t), t)
eta2_dd = diff(eta2_d, t)
```

### Solve for $\eta_2$

```python
eq1 = eta2_dd + omega0**2*eta2(t) + 2*beta*eta2_d - A*cos(omega*t)
eq1
```

```python
sol1 = dsolve(eq1)
sol1
```

### Solve for $\eta_1$

```python
eq2 = eta1_dd + omega0**2*eta1(t) + 2*beta*eta1_d + 4*omega0**2*epsilon*eta1(t) - A*cos(omega*t)
eq2
```

```python
sol2 = dsolve(eq2)
sol2
```
