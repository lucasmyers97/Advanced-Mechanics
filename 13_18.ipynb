# Phase velocity and group velocity for loaded string

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
dpi = 180
mpl.rcParams['figure.dpi'] = dpi
```

## Define functions for phase and group velocity

```python
def U(k):
    return np.cos(k)

def V(k):
    return np.abs(np.sin(k))/k
```

```python
fig, ax = plt.subplots(1)
k = np.linspace(0, np.pi/2, num=1000)

ax.plot(k, V(k), label="$\overline{V}(\overline{k})$")
ax.plot(k, U(k), label="$\overline{U}(\overline{k})$")
ax.legend()

ax.set_title("Phase and group velocities vs. wave number")
ax.set_xlabel("$\overline{k}$")
ax.set_ylabel("Velocity")

filename = "PhaseGroupVelocities.png"
fig.savefig(filename)
```
