# String with one section more dense

```python
from sympy import *
init_printing()
```

```python
A, B, C, D, E, k1, k2, L = symbols('A B C D E k_1 k_2 L', real=True)
```

## Solve for $B, C, D, E$ in terms of $A$

```python
expr1 = A + B - (C + D)
expr2 = B - A - ((k2/k1)*(D - C))
expr3 = C*exp(I*(k1 - k2)*L) + D*exp(I*(k1 + k2)*L) - E
expr4 = (k2/k1)*(C*exp(I*(k1 - k2)*L) - D*exp(I*(k1 + k2)*L)) - E
```

```python
sol = simplify(linsolve([expr1, expr2, expr3, expr4], (E, D, C, B, A)))
display(sol.args[0])
```

```python
Esol = sol.args[0][0]
Dsol = sol.args[0][1]
Csol = sol.args[0][2]
Bsol = sol.args[0][3]
Asol = sol.args[0][4]

display(Esol, Dsol, Csol, Bsol, Asol)
```

## Figure out what complex square of denominator is

```python
display((expand(Esol.conjugate()*Esol)))
```
