# Gaussian Wave Packet

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

dpi = 180
mpl.rcParams['figure.dpi'] = dpi
```

## Plot Gaussian wave packet

```python
del(k0)
```

```python
def psi(x, k0):
    return np.exp(-x**2)*np.exp(-1j*x*k0)
```

```python
fig, ax = plt.subplots(2, 2)
x = np.linspace(-2*np.pi, 2*np.pi, num=1000)

k0_ar = [1, 5, 10, 20]

for i in range(2):
    for j in range(2):
        k0 = k0_ar[2*i + j]
        ax[i][j].plot(x, np.real(psi(x, k0)))
        
        ax[i][j].set_title("$\overline{k_0} = $" + str(k0))
        ax[i][j].set_xlabel("$\overline{x}$")
        ax[i][j].set_ylabel("$\overline{\Psi}$")
        
fig.tight_layout()
filename = "GaussianPackets.png"
fig.savefig(filename)
```

## Show spreading out

```python
def packet(x, t):
    return np.exp(-(t - x)**2/(4 + 4*t**2))*np.sqrt(2*np.pi/(2 - 1j*t))
```

```python
fig, ax = plt.subplots(1)

x = np.linspace(-20, 20, num=1000)
t_ar = [0, 1, 3, 5]
for t in t_ar:
    ax.plot(x, np.real(packet(x + t, t)), label="$t = $" + str(t))
    
ax.legend()
ax.set_title("Widening of Gaussian Wave Packet")
ax.set_xlabel("x + t")
ax.set_ylabel("Amplitude")

filename = "WideningGaussian.png"
fig.savefig(filename)
```
