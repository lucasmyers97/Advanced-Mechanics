# String given initial velocity on middle segment

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.animation import FuncAnimation
```

```python
%matplotlib inline
dpi = 180
mpl.rcParams['figure.dpi'] = dpi
```

## Define $\overline{q}(\overline{x}, \overline{t})$

```python
def q(x, t, n, s):
    return -sum( (1j/(r*np.pi)**2)*( np.sin(r*np.pi/2)*
                                     np.sin(r*np.pi*s)*
                                     np.sin(r*np.pi*x)*
                                     np.exp(1j*r*np.pi*t) )
                                    
                for r in range(1, n + 1))
```

```python
fig, ax = plt.subplots(1)

n = 1000
s = 1/5

x = np.linspace(0, 1, num=1000)

t_ar = np.linspace(0, 1, num=12)
for t in t_ar:
    plot_label = "$\overline{t} = $" + str(round(t, 2))
    ax.plot(x, np.real(q(x, t, n, s)), label=plot_label)
    
ax.legend()
ax.set_title("$\overline{q}$ vs. $\overline{x}$ for various $\overline{t})$")
ax.set_ylabel("$\overline{q}(\overline{x}, \overline{t})$")
ax.set_xlabel("$\overline{x}$")
```

```python
fig, ax = plt.subplots(1)
ln, = ax.plot([], [])

n = 1000
s = 1/100
x = np.linspace(0, 1, num=1000)

ax.set_title("$\overline{q}$ vs. $\overline{x}$ for $s_0 = $" + str(round(s, 2)))
ax.set_ylabel("$\overline{q}(\overline{x}, \overline{t})$")
ax.set_xlabel("$\overline{x}$")
ax.set_xlim(0, 1)
ax.set_ylim(-0.08, 0.08)

def update(t, ln, x, q, n, s):
    ln.set_data(x, np.real(q(x, t, n, s)))
    return ln,
    
    
t = np.linspace(0, 2, num=1000)
ani = FuncAnimation(fig, update, t, fargs=(ln, x, q, n, s), blit=True, interval=10)
```

```python
ani.save("PluckedStringSmallSection.mp4")
```
