# Comparing discrete and continuous characteristic frequencies

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
dpi = 180
mpl.rcParams['figure.dpi'] = dpi
```

## Define functions for $\omega_r$ for continuous and discrete cases

```python
def omega(r, n):
    if n is not 0:
        return 2*(n + 1)*np.sin(r*np.pi/(2*(n + 1)))
    else:
        return r*np.pi
```

```python
r = np.linspace(0, 11*np.pi, num=1000)

fig, ax = plt.subplots(1)

n_ar = [3, 5, 10, 0]
for n in n_ar:
    
    if n is not 0:
        plot_label = "$n = $" + str(n)
    else:
        plot_label = "$n = \infty$"
        
    ax.plot(r, omega(r, n), label=plot_label)
    
ax.set_title("$\overline{\omega}_r$ vs. $r$")
ax.set_xlabel("$r$")
ax.set_ylabel("$\overline{\omega}_r$")
ax.legend()

filename = "CharacteristicFrequencies.png"
fig.savefig(filename)
```
